```python
def fix_path(old_path):
    if '/data/medical/' in old_path or 'data/medical/' in old_path:
        filename = old_path.split('/')[-1]
        return f'/content/drive/MyDrive/dir_dataset/data/medical/{filename}'
    elif '/data/non-medical/' in old_path or 'data/non-medical/' in old_path:
        filename = old_path.split('/')[-1]
        return f'/content/drive/MyDrive/dir_dataset/data/non-medical/{filename}'
    else:
        return old_path  # fallback: leave unchanged if pattern not found

diarize_dataset['path'] = diarize_dataset['path'].apply(fix_path)
```

```python
# prompt: i want to loop through the diarize_dataset dataframe and go the path column and loop through the 33 rows and run diarization in the 33 paths and store the result in a new colum called pred_segment

diarize_dataset['pred_segment'] = ''

for index, row in diarize_dataset.iterrows():
  audio_path = row['path']  # Assuming 'path' is the column with audio file paths
  try:
    # Run diarization on the audio file
    result = sd_pipeline(audio_path)
    # Store the result in the new 'pred_segment' column
    diarize_dataset.at[index, 'pred_segment'] = result
  except Exception as e:
    print(f"Error processing {audio_path}: {e}")
    diarize_dataset.at[index, 'pred_segment'] = None # Or some indicator of failure

print(diarize_dataset.head())
```

```python
import re

def time_str_to_sec(time_str):
    """Convert HH:MM:SS to seconds."""
    h, m, s = [int(x) for x in time_str.strip().split(':')]
    return h * 3600 + m * 60 + s

def parse_transcript_custom(transcript):
    lines = [l.strip() for l in transcript.strip().splitlines() if l.strip()]
    segments = []
    times = []
    speakers = []
    # Collect all timestamps and speakers with utterance order
    for i, line in enumerate(lines):
        # Only process lines that start with the timestamp pattern
        if re.match(r'^\d{2}:\d{2}:\d{2}', line):
            times.append(time_str_to_sec(line.split(' ')[0])) # Extract only the timestamp part
            # Check if speaker information is on the same line after timestamp
            if re.search(r'\[Speaker \d+\]:', line):
                 m = re.search(r'\[Speaker (\d+)\]:', line)
                 speaker = int(m.group(1)) - 1  # make 0-based for pyannote
                 speakers.append((len(times)-1, speaker)) # (timestamp index, speaker)
        elif re.match(r'\[Speaker \d+\]:', line) and not times:
             # Handle cases where the transcript starts with a speaker tag without a preceding timestamp
             m = re.match(r'\[Speaker (\d+)\]:', line)
             speaker = int(m.group(1)) - 1  # make 0-based for pyannote
             # Add a default start time of 0 if no timestamp is found
             times.append(0)
             speakers.append((len(times)-1, speaker))


    # Build segments
    for i, (time_idx, speaker) in enumerate(speakers):
        start = times[time_idx]
        # Find the next speaker tag timestamp
        next_start = None
        for j in range(i + 1, len(speakers)):
            next_time_idx, _ = speakers[j]
            if next_time_idx < len(times):
                next_start = times[next_time_idx]
                break

        if next_start is not None:
            end = next_start
        else:
            # If no subsequent speaker tag, use the end of the audio duration as the end time
            # This requires access to the audio_duration, which is not available in this function scope.
            # A simpler fallback is to add a default duration, but using the actual audio duration is better.
            # For now, we'll use a default fallback.
            end = start + 3  # fallback: add 3 seconds if no end time found

        segments.append([start, end, speaker])
    return segments
```

```python
import re

def parse_pred_segment_readable(pred_str):
    segments = []
    for line in pred_str.strip().splitlines():
        m = re.match(r"Speaker (\d+): ([\d.]+)s - ([\d.]+)s", line)
        if m:
            speaker = int(m.group(1)) - 1  # 0-based for pyannote
            start = float(m.group(2))
            end = float(m.group(3))
            segments.append([start, end, speaker])
    return segments
```

```python
import re

def time_str_to_sec(time_str):
    """Extract and convert HH:MM:SS to seconds."""
    match = re.match(r'(\d{2}):(\d{2}):(\d{2})', time_str.strip())
    if not match:
        return None
    h, m, s = map(int, match.groups())
    return h * 3600 + m * 60 + s

def parse_transcript_custom(transcript):
    lines = [l.strip() for l in transcript.strip().splitlines() if l.strip()]
    segments = []
    times = []
    speakers = []
    # Only pick lines that match a time
    for i, line in enumerate(lines):
        if re.match(r'^\d{2}:\d{2}:\d{2}$', line):
            t = time_str_to_sec(line)
            if t is not None:
                times.append(t)
        elif re.match(r'^\[Speaker \d+\]:', line):
            m = re.match(r'^\[Speaker (\d+)\]:', line)
            speaker = int(m.group(1)) - 1
            speakers.append((len(times)-1, speaker))
    for i, (time_idx, speaker) in enumerate(speakers):
        if time_idx < 0 or time_idx >= len(times):
            continue  # skip if index doesn't match
        start = times[time_idx]
        end = times[time_idx+1] if (time_idx+1 < len(times)) else start+3
        segments.append([start, end, speaker])
    return segments
```

```python
from pyannote.core import Annotation, Segment
from pyannote.metrics.diarization import DiarizationErrorRate

def to_pyannote_annotation(segments):
    annotation = Annotation()
    for seg in segments:
        start, end, speaker = seg
        annotation[Segment(start, end)] = str(speaker)  # pyannote expects speaker labels as str
    return annotation
  # Your pred_segments_struct

der = DiarizationErrorRate()

ders = []
for idx, row in diarize_dataset.iterrows():
    ref_segments = row['transcript_segments']
    hyp_segments = row['pred_segments_struct']
    if not ref_segments or not hyp_segments:
        continue
    ref_ann = to_pyannote_annotation(ref_segments)
    hyp_ann = to_pyannote_annotation(hyp_segments)
    ders.append(der(ref_ann, hyp_ann))
print(f"Average DER: {sum(ders)/len(ders):.2%}")




```
